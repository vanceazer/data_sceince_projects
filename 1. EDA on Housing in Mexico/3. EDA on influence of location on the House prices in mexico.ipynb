```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/mexico-real-estate-clean.csv')
df
```

```python
# with regards to location, which state has the highest price for housing
# calculating the mean house price grouped by state

mean_price_state = df.groupby('state')['price_usd'].mean().sort_values(ascending=False)
mean_price_state
```

```python
# visualizing this mean price in a bar chart
mean_price_state.plot(
    kind = 'bar',
    xlabel = 'state',
    ylabel = 'Prince in USD',
    title = 'Mean House Price by State'
);
```

```python
# another metric to measure can be the price per m2. 
# this will give give insight to the true value of a house based on location
df['price_per_m2'] = df['price_usd']/df['area_m2']
df
```

```python
# using this new metrics lets see how this looks on a bar chart
# first the mean grouped by state
mean_price_per_m2_by_state = df.groupby('state')['price_per_m2'].mean().sort_values(ascending=False)
mean_price_per_m2_by_state
```

```python
# bar chart for price per m2 by state
mean_price_per_m2_by_state.plot(
    kind = 'bar',
    xlabel = 'state',
    ylabel = 'Prince in USD',
    title = 'Mean House Price per m2 by State'
);
```
