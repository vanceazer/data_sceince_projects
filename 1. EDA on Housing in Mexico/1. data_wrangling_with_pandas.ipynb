```python
import pandas as pd

# with focus on mexico real estate data 1
df1 = pd.read_csv('data/mexico-real-estate-data-1.csv', encoding='latin-1')
df1.info()
df1.head()
```

```python
# dropping the rows that contain the NaN values
df1.dropna(inplace=True)
df1
```

```python
# working on the price_usd column to remove the $ and the comma and make it a float
df1['price_usd'] = (df1['price_usd'].str.replace("$","",regex=False)
                    .str.replace(",","")
                    .astype(float)
                   )

df1['price_usd']
```

```python
df1.info()
```

```python
# with focus on mexico real estate data 2
df2 = pd.read_csv('data/mexico-real-estate-data-2.csv', encoding='latin-1')
df2.info()
df2.head()
```

```python
# dropping the rows that contain the NaN values
df2.dropna(inplace=True)
df2
```

```python
# converting the pesos to usd on the price_mxn column
df2['price_usd'] = (df2['price_mxn']/19).round(2)
df2
```

```python
# dropping the price in mxn colum
df2.drop(columns=['price_mxn'], inplace=True)
df2.head()
```

```python
# with focus on mexico real estate data 3
df3 = pd.read_csv('data/mexico-real-estate-data-3.csv', encoding='latin-1')
df3.info()
df3.head()
```

```python
# separating the lat lon into different tables and dropping the Nan rows
df3.dropna(inplace=True)

df3[["lat","lon"]] = df3['lat-lon'].str.split(",", expand=True)
df3
```

```python
# splitting the place with parent names and extractinog the state
df3['state'] = df3['place_with_parent_names'].str.split('|', expand=True)[2]
df3
```

```python
# dropping unwanted columns
df3.drop(columns = ['place_with_parent_names', 'lat-lon'], inplace=True)
df3
```

```python
# combining all the cleaned dataframes into one dataframe
combined_df = pd.concat([df1,df2,df3])
print(combined_df.shape)
combined_df.info()
combined_df.head()
```

```python
# save the new df to a csv
combined_df.to_csv('data/mexico-real-estate-clean.csv', index=False)
```
