```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/mexico-real-estate-clean.csv')
df
```

```python
# checking to see if there is a relationship between the homesize and price
# using a scatter plot
plt.scatter(x=df['area_m2'], y=df['price_usd'])
plt.xlabel('Area [sq meters]')
plt.ylabel('Price [USD]')
plt.title('Price vs Area');
```

```python
# calculating the pearson correlation coefficient to measure the linear correlation
p_corr = df['area_m2'].corr(df['price_usd']).round(2)
p_corr
```

```python
# the 0.59 correltaion indicates there is a positive increase of proportion and it beind at 0.59 indicates that its moderate
```

```python
# focusing on a state of interest like Morelos to see the relationship there
# create a subset dataframe focuses on only Morelos

df_morelos = df[df['state'] == 'Morelos']
df_morelos
```

```python
# visualizaing the data in a scatter plot
plt.scatter(x=df_morelos['area_m2'], y=df_morelos['price_usd'])
plt.xlabel('Area [sq meters]')
plt.ylabel('Price [USD]')
plt.title('Morelos: Price vs Area');
```

```python
# calculationg the pearson coefficient
p_corr_morales = df_morelos['area_m2'].corr(df_morelos['price_usd']).round(2)
p_corr_morales
```

```python
# this indicates a strong relationship between house sizes and price in Morales
```
